```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
import matplotlib.colors as colors
import seaborn as sns
import pandas as pd
```

```python
time = []
n_2D_11e11_250K = []
n_2D_11e11_125K  = []
n_2D_11e11_83K  = []
n_2D_73e11_250K = []
n_2D_73e11_125K  = []
n_2D_73e11_93K  = []
n_2D_11e12_250K = []
n_2D_11e12_125K  = []
n_2D_11e12_114K  = []
n_2D_15e12_250K = []
n_2D_15e12_125K  = []
n_2D_15e12_102K  = []
with open('Sheet1.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        time.append(float(row[0]))
        n_2D_11e11_250K.append(float(row[1]))
        n_2D_11e11_125K.append(float(row[2]))
        n_2D_11e11_83K.append(float(row[3]))
        n_2D_73e11_250K.append(float(row[4]))
        n_2D_73e11_125K.append(float(row[5]))
        n_2D_73e11_93K.append(float(row[6]))
        n_2D_11e12_250K.append(float(row[7]))
        n_2D_11e12_125K.append(float(row[8]))
        n_2D_11e12_114K.append(float(row[9]))
        n_2D_15e12_250K.append(float(row[10]))
        n_2D_15e12_125K.append(float(row[11]))
        n_2D_15e12_102K.append(float(row[12]))
        
```

```python

log_proper250 = np.log(n_2D_11e11_250K)
print(log_proper250)
log_p1_0s = np.polyfit(time, log_proper250,1)

log_proper83 = np.log(n_2D_11e11_125K)
print(log_proper83)
log_p2_0s = np.polyfit(time, log_proper83,1)

log_proper25012 = np.log(n_2D_15e12_250K)
print(log_proper25012)
log_p3_0s = np.polyfit(time, log_proper25012,1)

log_proper102 = np.log(n_2D_15e12_125K)
print(log_proper102)
log_p4_0s = np.polyfit(time, log_proper102,1)
```
