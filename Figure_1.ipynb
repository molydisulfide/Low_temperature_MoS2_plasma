```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
from matplotlib import pyplot
import matplotlib.colors as colors
import bokeh as bk
import seaborn as sns
import pandas as pd
```

```python
xvth = []
vth_0s = []
vth_0s_err = []
vth_4s = []
vth_4s_err = []
vth_6s = []
vth_6s_err = []
vth_8s = []
vth_8s_err = []
with open('V_ths.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        xvth.append(float(row[0]))
        vth_0s.append(float(row[1]))
        vth_0s_err.append(float(row[2]))
        vth_4s.append(float(row[3]))
        vth_4s_err.append(float(row[4]))
        vth_6s.append(float(row[5]))
        vth_6s_err.append(float(row[6]))
        vth_8s.append(float(row[7]))
        vth_8s_err.append(float(row[8]))
```

```python
fig = plt.figure(figsize=(2,4))

fig.canvas.draw()

ax1 = fig.add_subplot(111)
ax1.plot(xvth,vth_0s, 'o', ms=7, color = 'g', lw=2, alpha=0.7, mfc='g', dash_capstyle = 'butt', mew = '1', label = '0 sec' ) 
ax1.plot(xvth,vth_4s, 's', ms=7, color = 'r',  lw=2, alpha=0.7, mfc='red', dash_capstyle = 'butt', mew = '1', 
              label = '4 sec' ) 
ax1.plot(xvth,vth_6s, '^', ms=7, color = 'b',  lw=2, alpha=0.7, mfc='b', dash_capstyle = 'butt', mew = '1', 
              label = '6 sec' ) 
ax1.plot(xvth,vth_8s, 'v', ms=7, color = 'black', lw=2, alpha=0.7, mfc='black', dash_capstyle = 'butt', mew = '1', label = '8 sec' ) 
 
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
ax1.set_ylabel('Threshold voltage (V)', fontsize = 12)
ax1.set_xlabel('Temperature (K)', fontsize = 12)
ax1.legend(loc= 'upper right',ncol = 1, fontsize = 10)
plt.savefig("Fig_1f.png",bbox_inches='tight', dpi = 300)


plt.show()
```
